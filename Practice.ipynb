```python
import numpy as np
import matplotlib.pyplot as plt

n = 500

x = np.random.uniform(0, 100, n)

a = 2
b = 20

y = a * x + b

noise = np.random.normal(0, 10, n) 

y_noisy = y + noise

 
plt.scatter(x, y_noisy, label="Random data with noise")
plt.plot(x, y, color="red", label="y = 2 * x + 20")
plt.legend()
plt.show()
```
