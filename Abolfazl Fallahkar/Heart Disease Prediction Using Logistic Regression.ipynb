```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# بارگذاری فایل سی اس وی که دانلود کردم
disease_df = pd.read_csv("framingham.csv")

# حذف ستون education و تغییر نام ستون male
disease_df.drop(columns=['education'], inplace = True, axis = 1)
disease_df.rename(columns ={'male':'Sex_male'}, inplace = True)
```

```python
# حذف ردیف‌های که داده ای ندارن برای سادگی
disease_df.dropna(axis = 0, inplace = True)
disease_df
```

```python
# نمایش توزیع کلاس هدف
print(disease_df.TenYearCHD.value_counts())
```

```python
# انتخاب ویژگی‌ها و برچسب هدف
# ویژگی‌های انتخاب‌شده بر اساس مقاله: سن، جنسیت، سیگار در روز، کلسترول، فشار خون سیستولیک، گلوکز
X = np.asarray(disease_df[['age', 'Sex_male', 'cigsPerDay', 
                           'totChol', 'sysBP', 'glucose']])
y = np.asarray(disease_df['TenYearCHD'])

# مقیاس‌بندی ویژگی‌ها با StandardScaler
# این کار برای Logistic Regression ضروری است تا ویژگی‌ها وزن یکسانی داشته باشند
X = preprocessing.StandardScaler().fit(X).transform(X)

# واردات train_test_split برای تقسیم داده‌ها
from sklearn.model_selection import train_test_split

# تقسیم داده‌ها به train (70%) و test (30%)
# random_state=4 برای تکرارپذیری نتایج
X_train, X_test, y_train, y_test = train_test_split( 
        X, y, test_size = 0.3, random_state = 4)

# و # نمایش اندازه مجموعه‌های ترین و تست
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
# رسم تعداد پلات ها برای نمایش توزیع کلاس هدف در کل دیتاست

plt.figure(figsize=(7, 5))
sns.countplot(x='TenYearCHD', data=disease_df,
             palette="BuGn_r")

# این ویژوالیزیشن نشان‌دهنده غیر بالانس بودن شدید کلاس‌ها است
plt.show()
```

```python
# رسم توزیع کلاس هدف با مت‌پلات‌لیب (نمودار ساده)
laste = disease_df['TenYearCHD'].plot()
plt.show(laste)
```

```python
# واردات مدل Logistic Regression
from sklearn.linear_model import LogisticRegression

# ساخت و آموزش مدل
# LogisticRegression یکی از ساده‌ترین و قابل‌تفسیرترین مدل‌ها برای طبقه‌بندی باینری است
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# پیش‌بینی روی داده‌های تست
y_pred = logreg.predict(X_test)
```

```python
# محاسبه و نمایش دقت مدل
from sklearn.metrics import accuracy_score
print('Accuracy of the model is =', 
      accuracy_score(y_test, y_pred))
```

```python
# نمایش گزارش کامل classification (precision, recall, f1-score)
# به دلیل imbalance، recall کلاس ۱ پایین است که نکته مهمی برای بحث در ارائه است
from sklearn.metrics import confusion_matrix, classification_report

print('The details for confusion matrix is =')
print (classification_report(y_test, y_pred))


# محاسبه confusion matrix و نمایش آن با heatmap
# heatmap خوانایی بهتری نسبت به ماتریس خام دارد
cm = confusion_matrix(y_test, y_pred)
conf_matrix = pd.DataFrame(data = cm, 
                           columns = ['Predicted:0', 'Predicted:1'], 
                           index =['Actual:0', 'Actual:1'])

plt.figure(figsize = (8, 5))
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens")

plt.show()
```
