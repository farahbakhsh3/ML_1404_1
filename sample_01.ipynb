```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
x = [[1],[2],[3],[4],]
y = [1.3, 3.6, 4.3, 6.5]


x = np.array(x)
y = np.array(y)
```

```python
x , y
```

```python
x.shape, y.shape
```

```python
plt.plot(x, y, '*')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
```

```python
model.coef_[0], model.intercept_
```

```python
model.predict([[10]])
```

```python
plt.plot(x, y, '*')
plt.plot(x, model.predict(x))
plt.show()
```

```python
x = np.linspace(1, 360, 360 )
y = np.array([2 * i + (np.random.rand() - 0.5) * 20 for i in x])

x = x.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
plt.plot(x, y)
plt.plot(x, model.predict(x))
plt.show()
```

```python
x[0], y[0], model.predict(x[0].reshape(-1, 1))
```

```python
x[3], y[3], model.predict(x[3].reshape(-1, 1))
```

```python
error = np.array([abs(y[i] - model.predict(x[i].reshape(-1, 1))) for i in range(1, 100)])
mae = np.average(error)

print(mae)
```

```python
# Generate x values
x = np.linspace(0, 4*np.pi, 200)

# Generate sin values with noise
y = np.sin(x) + np.random.normal(0, 0.1, len(x))

# Create the plot
plt.plot(x, y, 'b.', label='sin(x) + noise', markersize=5)
plt.show()
```

```python
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

model = LinearRegression()
model.fit(x, y)

plt.plot(x, y, '.')
plt.plot(x, model.predict(x), '--')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(max_depth=5)
model.fit(x, y)

plt.plot(x, y, '.')
plt.plot(x, model.predict(x), '--')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor as m

model = m(n_neighbors=5)
model.fit(x, y)

plt.plot(x, y, '.')
plt.plot(x, model.predict(x), '--')
plt.show()
```
