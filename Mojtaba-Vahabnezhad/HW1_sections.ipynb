## **1. Importing Libraries**


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## **2. Creating a Random Linear Dataset**



```python
x = np.linspace(1, 200, 200)
y = np.array([3*i + (np.random.rand() - 0.5) * 40 for i in x])

x = x.reshape(-1, 1)

x[:5], y[:5]
```

## **3. Fitting the Linear Regression Model**


```python
model = LinearRegression()
model.fit(x, y)

print('Coefficient:', model.coef_[0])
print('Intercept:', model.intercept_)
print('Prediction for x = 50:', model.predict([[50]])[0])
```

## **4. Plotting the Data and Regression Line**


```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, '.', label='Data')
plt.plot(x, model.predict(x), '--', label='Linear Regression')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.show()
```

## **5. Calculating MAE (Mean Absolute Error)**


```python
error = np.array([
    abs(y[i] - model.predict(x[i].reshape(1, -1)))
    for i in range(len(x))
])

mae = np.average(error)
print('MAE:', mae)
```
