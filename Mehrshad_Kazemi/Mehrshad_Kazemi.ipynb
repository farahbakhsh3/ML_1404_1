```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#sakht Data
x = np.arange(1, 501)  # 1 to 500
noise = np.random.uniform(1, 200, 500)
y = 2 * x + noise
x = x.reshape(-1, 1)

#train
model = LinearRegression()
model.fit(x, y)

#shib khat
slope = model.coef_[0]  
intercept = model.intercept_

print("Slope:", slope)
print("Intercept:", intercept)

#predict va subset
subset_start = 200
subset_end = 300
x_subset = x[subset_start:subset_end]
y_subset_pred = model.predict(x_subset)

#plot ba axline
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, y, '*', markersize=3, label="All Data")


#regression khati ba axline
ax.axline((0, intercept), slope=slope, color='green', linestyle='-', linewidth=2,
          label=f"Model Line: y={slope:.2f}x + {intercept:.2f}")
ax.plot(x_subset, y_subset_pred, linewidth=2, label="Subset Prediction")

ax.legend()
plt.show()
```
