```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

# =======================================
#  500 
# =======================================
np.random.seed(42)

x1 = np.linspace(0, 500, 500)
y1 = 2 * x1 + (np.random.rand(len(x1)) - 0.5) * 100

X1 = x1.reshape(-1, 1)

plt.plot(x1, y1, '*')
plt.title("Linear Dataset (500 points)")
plt.show()

# Linear Regression   
model1 = LinearRegression()
model1.fit(X1, y1)

print("Linear Dataset -> coef:", model1.coef_[0])
print("Linear Dataset -> intercept:", model1.intercept_)
print("Predict(300):", model1.predict([[300]]))

plt.plot(x1, y1, '*')
plt.plot(x1, model1.predict(X1), '--')
plt.title("Linear Regression on Linear Data")
plt.show()


# =======================================
#  
# =======================================
x2 = np.linspace(0, 4*np.pi, 200)
y2 = np.sin(x2) + np.random.normal(0, 0.12, len(x2))

X2 = x2.reshape(-1, 1)

plt.scatter(x2, y2, s=10)
plt.title("Sinusoidal Data with Noise")
plt.show()


# =======================================
#Linear Regression  
# =======================================
linear = LinearRegression()
linear.fit(X2, y2)

plt.scatter(x2, y2, s=10)
plt.plot(x2, linear.predict(X2), 'r--')
plt.title("Linear Regression on Sin Data")
plt.show()

print("Sin -> Linear Coef:", linear.coef_[0])
print("Sin -> Linear Intercept:", linear.intercept_)


# =======================================
# 4) Decision Tree
# =======================================
tree = DecisionTreeRegressor(max_depth=6)
tree.fit(X2, y2)

plt.scatter(x2, y2, s=10)
plt.plot(x2, tree.predict(X2), 'g--')
plt.title("Decision Tree Regression")
plt.show()


# =======================================
# 5) KNN Regression
# =======================================
knn = KNeighborsRegressor(n_neighbors=7)
knn.fit(X2, y2)

plt.scatter(x2, y2, s=10)
plt.plot(x2, knn.predict(X2), 'm--')
plt.title("KNN Regression")
plt.show()
```
